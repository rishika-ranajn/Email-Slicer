```python
import pandas as pd

def slice_email(email):
    if "@" not in email:
        return None, None
    
    username, domain = email.split("@")
    return username, domain

def main():
    print("===== Email Slicer =====")
    
    emails = []
    
    while True:
        email = input("\nEnter email (or 'exit'): ")
        
        if email.lower() == "exit":
            break
        
        username, domain = slice_email(email)
        
        if username is None:
            print("Invalid email!")
            continue
        
        emails.append({"Email": email, "Username": username, "Domain": domain})
        print(f"Username: {username}, Domain: {domain}")
    
    # Convert to DataFrame
    df = pd.DataFrame(emails)
    
    if not df.empty:
        print("\n📊 All Emails Data:")
        print(df)

main()
```
